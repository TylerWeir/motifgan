# Data Preprocessing Pipeline
Used to prepare audio samples for training a DCGAN.

```python
import os
from pydub import AudioSegment
import matplotlib.pyplot as plt
import librosa
from IPython.display import Audio
```

### Read and plot the clip lengths

```python
# Source File path
source = "./samples/"

durations = []

# Read the duration of each clip in the source file
for i, sample in enumerate(os.listdir(source)):
    print(source + sample)
    try:
        # Try to read the sample duration
        song = AudioSegment.from_mp3(source + sample)
        durations.append(song.duration_seconds)
    except:
        # Delete the sample if there is an error
        os.remove(source+sample)
```

```python
# Plot Histogram on x
plt.hist(durations, bins=100)
plt.gca().set(title='Lengths of Clips', ylabel='Number of Clips', xlabel='Duration (Seconds)');
```

```python
#Scale clips to be about 3 seconds
target_duration  = 3

# Source File path
source = "./samples/475.mp3"

# Read the duration of each clip in the source file
#for i, sample in enumerate(os.listdir(source)):
    
# Try to read the sample duration
song = AudioSegment.from_mp3(source)
sample_dur = (song.duration_seconds)
    
scale_factor = target_duration / sample_dur
    
# time stretch by 1/scale_factor to get the target sample length
y, sr = librosa.load(source)
y_adjusted = librosa.effects.time_stretch(y, rate=1/scale_factor)
    
Audio(data=y_adjusted, rate=sr)
    
```
